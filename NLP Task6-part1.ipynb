### Tokenizing a sentence, then pos tagging each sentence and then using regexp parser

```python
para="""
Online shopping has become increasingly popular in recent years, and as a result, the volume of online reviews has also grown.
POS tagging is an important tool in NLP that can be used to analyze online reviews to gain insights into customer opinions and 
preferences. By identifying the parts of speech in online reviews, businesses can better understand what customers like and 
dislike about their products and services. For example, POS tagging can help identify specific adjectives that customers 
commonly use to describe a product, which can then be used to improve product descriptions and marketing strategies. In addition
, POS tagging can help identify patterns in customer feedback, such as common complaints or frequent compliments, which can be 
used to guide product development and customer service efforts. Overall, POS tagging is a powerful tool for businesses looking 
to improve their online presence and 
customer satisfaction."""
```

```python
from nltk.tokenize import sent_tokenize
sent=sent_tokenize(para)
```

```python
from nltk.tokenize import word_tokenize

tokens=[word_tokenize(i) for i in sent]
```

```python
#tokens
```

```python
tag=[nltk.pos_tag(i) for i in tokens]
```

```python
tag[0]
```

#### determiner,adverb,adjective,noun--->noun phrase

```python
grammar="NP:{<DT>?<JJ>*<NN|NNS|NNP.?>}"
```

```python
cp=nltk.RegexpParser(grammar)
```

```python
result=[cp.parse(i) for i in tag]
```
